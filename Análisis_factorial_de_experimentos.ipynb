#**Análisis de experimentos**

 Nuestro caso hipotético será el de una chef que adora hacer cupcakes y le salen muy bien, por lo que decide emprender un negocio de venta de cupcakes.

 Pero el desafío que enfrenta es que ella va a tener que producir gran cantidad cupcakes y a la vez intentar reducir al mínimo los gastos para producir esa cantidad con el fin de obtener beneficio económico.

 Nos indicó el número de ingredientes que generan una cantidad de cupcakes específica. Lo que más le preocupa a ella de todos los ingredientes es la harina y la leche, por ser los más caros. Vamos a ayudarla a planificar mejor sus experimentos, ella va a tener que ir variando la cantidad de harina y la cantidad de leche en sus recetas e ir observando o anotando cuántos cupcakes va produciendo en cada uno de estos experimentos.

 A través de un modelo estadístico, nosotros vamos a poder generar ecuaciones o modelos que aprendan muy bien cómo influye la harina y la leche en la receta de la chef.

 La idea es conseguir pronosticar correctamente la cantidad de cupcakes que se puede producir con una determinada cantidad de harina, con una determinada cantidad de leche y también con una cantidad limitada de experimentos. Todo esto a través de modelos estadísticos, modelos de Machine Learning.

## Modelo estadístico a utilizar:

mod_estadistico1.png

Esta es la fórmula básica que usaremos para representar estadíscamente nuestra receta:  donde 'P' es la cantidad de cupcakes producidos, 'X' representa la cantidad de harina que se utilizó en la receta, 'Y' es la cantidad de leche usada, y aquí vamos a adicionarle un elemento más: la interacción de la harina con la leche: X harina por Y leche, que adicionan un elemento o una variable que represente la interacción de la harina y la leche.



mod_estadistico2.png




Vamos a transformar esta fórmula básica en una fórmula un poco más avanzada para generar un análisis estadístico:

Aquí  'P' es la cantidad de cupcakes producidos, el primer elemento es la variable (B0)  que es el interceptor, que representa la cantidad de cupcakes producidos en total por los 4 experimentos divivido entre 4 (el número de experimentos).

El segundo elemento representa la cantidad de harina individual que estamos usando en la receta(B1) y la cantidad de leche individual que estamos usando en la receta(B2), estos dos variables individuales en la receta los vamos a llamar **efectos aislados**, ya que se cuentan por sí solos, no tienen ninguna interacción.

El tercer elemento es la interacción, de la harina con la leche, lo vamos a representar con un (B3), cuando mezclamos dos elementos en una fórmula química, en una receta, estos dos elementos se combinan y generan algo nuevo al cual lo vamos a llamar el **efecto de interacción**.

Y por último tenemos el error, que se representa con un (E), este símbolo error significa los experimentos que no consiguieron reproducir el ensayo, que no obtuvieron ninguna conclusión.

A través de este modelo estadístico podemos, analizar los experimentos que hemos obtenido en nuestro DataFrame, luego podremos analizar la significancia estadística que quiere decir la importancia estadística de estos cuatro parámetros, B0, B1, B2, B3 y el error, y así poder identificar si alguno de estos es importante o no, para poder retirarlo de nuestra ecuación y así poder ajustar nuestro modelo para obtener realmente la cantidad de cupcakes que estamos buscando al menor costo posible.

## Bibliotecas  Stats Model

Con nuestro modelo estadístico definido, ahora es necesario ahora saber los valores de cada uno de estos parámetros: el B0, el B1, el B2, el B3 y el error.

Estos parámetros están contenidos dentro de mis datos, que son los resultados de los cuatro experimentos que realizamos con el diseño factorial. Estos experimentos generaron una cantidad específica de cupcakes, cambiando o variando la cantidad de harina y de leche, luego de esos datos yo puedo extraer las informaciones de estos cuatro parámetros. Una vez que identifique cuáles son los valores de estos cuatro parámetros, voy a poder ajustar mi modelo y observar cuál de estos parámetros tiene significancia estadística.

Para obtener estos valores, Python cuenta con una librería específica : **Stats Model** que es especial para poder realizar análisis estadísticos, inferencias estadísticas y muchas otras aplicaciones estadísticas.

[statsmodels 0.14.0](https://www.statsmodels.org/stable/index.html)



```python
#Importando librerías
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
#Creando la fórmula del modelo estadístico
modelo = smf.ols(data=experimento, formula='Cantidad ~ Harina + Leche + Harina:Leche')#smf.ols=statsmodels.regression.linear_model.OLS
#A través del método ols, voy a poder extraer estos coeficientes del df 'experimento'
#Las variables interceptor (B0) Y error (E) las va a generar el mismo método ols, cantidad(P), Harina(B1),Leche(B2)
#y Harina:Leche(B3) se obtienen de los datos de los 4 experimentos.
```

```python
modelo_ajustado = modelo.fit()
print(modelo_ajustado.summary())
```

El modelo ha calculado un valor para el interceptor(B0) de 32.25.  
La segunda variable de la harina(B1) es de 10.75.  La tercera variable de la leche(B2) es 4.25. Y la última que es la interacción entre la harina y la leche(B3), recibió un valor de 1.75.

El error que está separado en otra columna nan, no fue identificado en los cuatro experimentos que obtuvimos, así que vamos a enfocarnos en estos cuatro valores.

Para ver si estos valores que muestra el modelo estadístico son importantes tenemos que hacer el análisis de
significación estadística que determinará si estos coeficientes son diferentes de 0 o no. Si son diferentes de cero significa que tienen alguna importancia, pero si alguno es igual a 0 podemos retirarlo de nuestra ecuación.

## Grados de libertad

Antes de realizar el análisis de significación estadística debemos revisar si la cantidad de experimentos en nuestro caso de estudio son suficientes, tenemos cuatro experimentos que los vamos a llamar **grados de libertad** que quiere decir la cantidad límite que tengo de experimentos en mi caso de estudio.

En el modelo estadístico que hemos definido, tenemos 4 parámetros: B0, B1, 2, B2, B3. Cada uno de estos parámetros va a consumir un grado de libertad por lo que, cuando usemos los 4 parámetros, nos quedaremos sin grados de libertad ya que estos también son 4.

No podremos verificar el valor del error estadístico en nuestro modelo, tampoco saber qué tan ajustado está, ni podremos utilizar el análisis de significación estadística porque nos quedaríamos con un grado de libertad residual igual a 0. ¿Qué es un grado de libertad residual?

Un **grado de libertad residual** ó **GLR**, es el número de experimentos menos la cantidad de parámetros, entonces tenemos cuatro experimentos, menos cuatro parámetros: 4 - 4 = 0.

mod_estadistico3.png

Esto quiere decir, que nuestro modelo y la cantidad de experimentos no son suficientes para poder realizar un análisis de significación estadística. Necesitamos más experimentos para que el GLR sea mayor a 0.

Vamos a hacer la evaluación de la significación estadística de una forma gráfica, para entender si un parámetro es significante o no. Para esto vamos a utilizar el valor de **t** calculada.

La idea es la misma que el anterior método que usamos. Vamos a aceptar o rechazar la hipótesis nula de cada uno de los parámetros. Si aceptamos la hipótesis nula significa que ese parámetro es no significante y si la rechazamos significa que ese parámetro sí es significante para nuestro modelo.Aqui la fórmula:

hipotesis.png





Cuando la probabilidad de **t** de algún parámetro era mayor o igual que alfa, entonces aceptábamos la hipótesis nula y ese parámetro no era significante. Recordar que el valor de alfa utilizado es de 5% porque es el estándar que se utiliza actualmente en estadística.
valort.png

Entonces vamos a coger ese valor de 5% y vamos a llevarlo a una distribución **t**. y luego lo dividimos en dos. Esta distribución t es una distribución simétrica, tiene la misma distancia de 0 a la izquierda como de 0 a la derecha, entonces cuando dividimos el valor de 5% en dos, va a ir la mitad al lado izquierdo, representando valores negativo 2.5%, y la otra mitad 2.5% representando los valores positivos.

Estos 2.5% en ambos lados va a generar un área que está sombreado de rojo, que le vamos a llamar de **t** tabulado. T tabulado va a definir si el límite máximo para que una variable consiga llegar al límite de significación. Observen aquí:
distribuciont.png

En el siguiente gráfico el **t** tabulado (**t tab**) en ambos lados,  está de color verde, luego colocamos el t calculado(**t calc**), que son los valores de color rosado, en el caso de que el **t calc** sea menor que el **t tab**, estará más a la izquierda, quiere decir que se acepta la hipótesis nula y por lo tanto esta variable no es significante.

Pero si el **t tab** es menor que el **t calc** querrá decir que se rechaza la hipótesis nula y el parámetro es significante para nuestra ecuación.


valid_hipot.png

Ahora haremos la prueba con el código

```python
t_valores = modelo_ajustado.tvalues# con t.values obtengo el valor de los parámetros
t_valores
```

```python
nombre = t_valores.index.to_list()# extraigo los índices(nombre) de los valores y los agrego a una lista
nombre
```

```python
from scipy import stats #método stats de la librería scipy para generar el t tabulado (t tab)
distribucion = stats.t(df=4)# guardamos el t tab en la variable distribución , df=GLR
distribucion.ppf( q= 1 - 0.025)
# El metodo ppf va a calcular el valor de t tabulado. Para eso tengo que pasarle como parámetro el valor de q.
# el valor de q es la diferencia del 100% menos el área que corresponde al t tabulado que es 2.5%,
# El 100 lo vamos a representar como 1, y 2.5% en formato decimal 0.025
```

```python
# El t tab es 2.77 entonces ese valor lo multiplicamos por el número de parametros del modelo para generar una línea que
# una el t tab para cada parámetro dentro del diagrama de pareto.
limite = [distribucion.ppf( q= 1 - 0.025)]* len(nombre)
limite
```

### Plotando el gráfico


```python
#Creamos nuestro diagrama de Pareto, cada barra representa el t calc de cada parámetro:
pareto = sns.barplot(x=t_valores, y=nombre)#eje x=valores de los parámetros, y=lista de nombres
pareto.figure.set_size_inches(15,6)#tamaño del gráfico
pareto.tick_params(labelsize=20)#tamaño de las etiquetas de t_valores(0,10,20,30,40)
pareto.set_xlabel('t_valores',fontsize=20)
pareto.plot(limite,nombre,'r');#línea que representa el valor t tab, nos indica si el parámetro es significante o no.
```

Ahora que ya tenemos nuestro modelo con los parámetros correctos, revisado, analizado y validado, debemos crear una interfaz amigable con el usuario final.

```python
# Obtenemos los parámetros de nuestro modelo que nos van a indicar su peso dentro de nuestra ecuación, es necesario saber
# esto para después poder calcular con nuevos valores de variables de harina y de leche.
parametros = modelo_ajustado_2.params
parametros
```

### Definiendo la función

Definimos la función en base al modelo actualizado:

mod_est_actual.png

```python
#El modelo utiliza valores normalizados, entre -1 y +1 que debemos convertir a valores numéricos comunes.
def modelo_receta(x_h,x_l):# x_h y x_l son los valores de harina y leche que va a colocar el usuario final
  #creamos los limites usando los valores normalizados del modelo para cada parámetro
  limite_normalizado = [-1,+1]
  limite_harina = [0.5,1.5]
  limite_leche = [0.2,0.7]
  #Tenemos que convertir el limite_harina y el limite_leche a una escala de -1 y +1.para eso usamos el metodo interp()
  #de Numpy
  x_h_convertido = np.interp(x_h,limite_harina, limite_normalizado)
  x_l_convertido = np.interp(x_l,limite_leche, limite_normalizado)

  #Ahora creamos la ecuación en base al modelo estadístico actualizado
  cantidades = parametros['Intercept']+ parametros['Harina']*x_h_convertido+ parametros['Leche']*x_l_convertido
  return round(cantidades)
```

```python
# En este ejemplo por 500 gr. de harina y 200 ml de leche se obtienen 16 cupcakes, y asi agregando valores distintos
# de harina y leche el modelo nos dirá cuantos cupcakes obtendremos
modelo_receta(0.5,0.2)
```

Vamos a generar una opción 2 de interfaz amigable. En este caso vamos a generar un mapa de colores. Ya habrán escuchando de mapa de calor, cuando más intenso el color, cuanto más rojo significa que existe una mayor cantidad de muestra en este sector, y cuanto más frío, cuanto más azul, una cantidad menor de muestras.

En nuestro caso para la cantidad de cupcakes. Cuanto más rojo, mayor cantidad de cupcakes que podemos producir, y cuanto más frío,  vamos a generar una menor cantidad de cupcakes con esta combinación de valores de la leche y de la harina.

De una forma gráfica va a utilizar esto como un diccionario de colores para que pueda saber la combinación exacta que necesita para su lote del día.

modelo_colores.png


Para realizar esto necesitamos generar la cantidad de cupcakes en todo el área de experimentación  viendo el límite superior y el límite inferior tanto de la harina como de la leche.

Los puntos negros en nuestra área de experimentación representan la cantidad de cupcakes producidos. Tenemos que calcular todas las cantidades de cupcakes producidas.

Una vez que generamos puntos para la harina y puntos para la leche que recorren toda el área de experimentación, vamos a llamar a nuestro, modelo y pedirle que nos calcule cuántos cupcakes serán producidos para cada combinación de puntos en toda el área y obtenemos estos puntos en negrita, que es la cantidad de cupcakes. Una vez obtenidos, vamos a transformarlo a colores y creamos nuestro mapa.

```python
#Generamos los puntos de la harina de una forma automática con el método linspace de Numpy
x_harina = np.linspace(start= 0.5, stop= 1.5, num=10)
x_harina
```

```python
#Creamos los puntos de la leche
x_leche = np.linspace(start= 0.2, stop= 0.7, num=10)
x_leche
```

```python
# Usamos los puntos o las cantidades de leche y la cantidad de harina, para crear un ciclo for anidado que calcule la cantidad de cupcakes
puntos= []# Array en blanco
for cont1 in x_harina:
  temp = []# para cada uno de los puntos de la harina vamos a generar una lista vacía temp
  for cont2 in x_leche:
    temp.append(modelo_receta(cont1,cont2))
  puntos.append(temp)#nos entrega la cantidad de cupcakes que predice el modelo
puntos
```

Estas son las cantidades de cupcakes en toda el area de experimentación, ahora vamos a asignar colores a estos puntos para poder visualizarlos en un gráfico
